# **Prompt Engineering**


````python
import os
from groq import Groq
from dotenv import load_dotenv, find_dotenv

_ = load_dotenv(find_dotenv())
````

````python
PROMPT = """Explain like I'm five years old: ```Consider these three strategies for prompt engineering: First, provide context. Second, be specific. And third, build on the conversation.```
"""
````

````python
client = Groq(api_key=os.environ.get("GROQ_API_KEY"))
````

````python
response = client.chat.completions.create(
    model="openai/gpt-oss-20b",
    temperature=0.2,
    messages=[
        {
            "role": "user",
            "content": PROMPT,
        }
    ],
)
````

````python
print(response.choices[0].message.content)
````
